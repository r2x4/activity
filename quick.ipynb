# **DataFrame de Tiempo**


```python
columnas_revisar = [
    "nombre_cliente",
    "documento_cliente",
    "creado_por",
    "estado",
    "novedad",
    "Conductor",
    "ventanas_horarias",
    "fecha_de_creacion",
    "fecha ciruguia",
    "asignado",
    "aceptado",
    "llegado",
    "atendido",
    "finalizado",
    "Municipio",
    "Departamento",
    "alcance_de_novedad"
]

# Usamos el DataFrame correcto: datos
df_eventos = datos[columnas_revisar].copy()

# Convertimos las columnas de fecha a tipo datetime
columnas_fecha = [
    "fecha_de_creacion",
    "fecha ciruguia",
    "asignado",
    "aceptado",
    "llegado",
    "atendido",
    "finalizado"
]

for col in columnas_fecha:
    df_eventos[col] = pd.to_datetime(df_eventos[col], errors='coerce')

df_eventos.head()

```

```python
# crear columnas para verificar tiempo

# Asegrate de que las columnas de fecha ya estn en formato datetime
df_eventos["asignado"] = pd.to_datetime(df_eventos["asignado"], errors='coerce')
df_eventos["aceptado"] = pd.to_datetime(df_eventos["aceptado"], errors='coerce')
df_eventos["finalizado"] = pd.to_datetime(df_eventos["finalizado"], errors='coerce')
df_eventos["atendido"] = pd.to_datetime(df_eventos["atendido"], errors='coerce')

# Calculamos el tiempo entre los eventos y lo asignamos al DataFrame
df_eventos["tiempo_asignado_aceptado"] = (df_eventos["aceptado"] - df_eventos["asignado"]).dt.total_seconds() / 3600  # en horas
df_eventos["tiempo_aceptado_finalizado"] = (df_eventos["finalizado"] - df_eventos["aceptado"]).dt.total_seconds() / 3600  # en horas
df_eventos["tiempo_atendido_finalizado"] = (df_eventos["finalizado"] - df_eventos["atendido"]).dt.total_seconds() / 3600  # en horas

# Mostrar las primeras filas para verificar
print(df_eventos[["tiempo_asignado_aceptado", "tiempo_aceptado_finalizado", "tiempo_atendido_finalizado"]].head())
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Configuracin de conexin a MySQL
usuario = 'root'
contrasena = 'Dragon2307*'
host = 'localhost'
puerto = '3306'
nombre_bd = 'quick'

# Crear motor de conexin
cadena_conexion = f"mysql+pymysql://{usuario}:{contrasena}@{host}:{puerto}/{nombre_bd}"
engine = create_engine(cadena_conexion)

# Clasificaciones de novedad
clasificaciones = {
    'SALIDA COMPLETA BODEGA': ('Entrega', 'Efectiva'),
    'RECOGIDA EFECTIVA': ('Recogida', 'Exitosa'),
    'REROUTE': ('Entrega', 'Efectiva'),
    'ENTREGA PARCIAL': ('Entrega', 'Efectiva'),
    'RECOGIDA NO EFECTIVA': ('Recogida', 'Fallidas'),
    'ENTREGA NO EFECTIVA': ('Entrega', 'No efectiva (Canceladas)'),
    'RECOGIDA PARCIAL': ('Recogida', 'Exitosa'),
    'RECOGIDA REMISION DIGITAL (FOTO)': ('Recogida', 'Exitosa'),
    'FALLA EN PLATAFORMA': ('Entrega', 'Picos operativos'),
    'H TO H': ('Entrega', 'Efectiva'),
    'ENTREGA EFECTIVA SIN NOVEDAD': ('Entrega', 'Efectiva')
}

# Funcin para clasificar novedad
def clasificar_novedad(novedad):
    return clasificaciones.get(novedad, ('Desconocido', 'Desconocido'))

# Aplicar clasificacin al DataFrame
datos['tipo_servicio'], datos['clasificacion_operacion'] = zip(*datos['novedad'].map(clasificar_novedad))

# --- Cargar tabla 'datos' sin duplicados ---
# Verificamos si existe la tabla
try:
    datos_existentes = pd.read_sql('SELECT guia FROM datos', con=engine)
    datos_nuevos = datos[~datos['guia'].isin(datos_existentes['guia'])]
    print(f" {len(datos_nuevos)} registros nuevos encontrados.")
except Exception:
    # Si la tabla no existe an, cargamos todo
    datos_nuevos = datos
    print(" Tabla 'datos' no existe, se cargar completamente.")

# Subir datos nuevos
if not datos_nuevos.empty:
    datos_nuevos.to_sql(name='datos', con=engine, if_exists='append', index=False)
    print(" Tabla 'datos' actualizada con nuevos registros.")
else:
    print(" No hay nuevos registros para agregar a la tabla 'datos'.")

# --- Subir tabla de eventos (se reemplaza cada vez) ---
df_eventos.to_sql(name='df_eventos', con=engine, if_exists='replace', index=False)
print(" Tabla 'df_eventos' actualizada.")

# --- Subir conductores nicos ---
df_conductores = datos[['Conductor', 'Documento del Conductor del Recurso']].drop_duplicates()
df_conductores.to_sql(name='conductores', con=engine, if_exists='replace', index=False)
print(" Tabla 'conductores' actualizada.")

# --- Subir clientes nicos ---
df_clientes = datos[['nombre_cliente', 'documento_cliente']].drop_duplicates()
df_clientes.to_sql(name='clientes', con=engine, if_exists='replace', index=False)
print(" Tabla 'clientes' actualizada.")

# --- Subir clasificaciones nicas ---
df_clasificaciones = datos[['tipo_servicio', 'clasificacion_operacion']].drop_duplicates()
df_clasificaciones.to_sql(name='clasificaciones', con=engine, if_exists='replace', index=False)
print(" Tabla 'clasificaciones' actualizada.")
```
